```python
# Finally, let's try mean pooling
def meanpool(x_in):
    x_out = np.zeros(
        (int(np.floor(x_in.shape[0] / 2)), int(np.floor(x_in.shape[1] / 2)))
    )
    height, width = x_in.shape[0] // 2 * 2, x_in.shape[1] // 2 * 2
    x_out = x_in[:height, :width]
    return x_out.reshape(height // 2, 2, width // 2, 2).mean(axis=(1, 3))
```

```python
# Define 2 by 2 original patch
orig_2_2 = np.array([[6, 8], [8, 4]])
print(orig_2_2)
```

```python
# Let's first use the duplication method
def duplicate(x_in):
    x_out = np.zeros((x_in.shape[0] * 2, x_in.shape[1] * 2))
    x_out = x_in.repeat(2, axis=0).repeat(2, axis=1)
    return x_out
```

```python
print("Original:")
print(orig_2_2)
print("Duplicated:")
print(duplicate(orig_2_2))
```

```python
# Now let's try max pooling back up
# The input x_high_res is the original high res image, from which you can deduce the position of the maximum index
def max_unpool(x_in: np.ndarray, x_high_res: np.ndarray):
    x_out = np.zeros((x_in.shape[0] * 2, x_in.shape[1] * 2))
    for i in range(x_in.shape[0]):
        for j in range(x_in.shape[1]):
            patch = x_high_res[2 * i : 2 * i + 2, 2 * j : 2 * j + 2]
            max_val_index = np.argmax(patch)
            offset_y, offset_x = np.unravel_index(max_val_index, patch.shape)
            x_out[2 * i + offset_y, 2 * j + offset_x] = x_in[i, j]
    return x_out
```

Finally, we'll try upsampling using bilinear interpolation. We'll treat the
positions off the image as zeros by padding the original image and round
fractional values upwards using np.ceil()


```python
def bilinear(x_in: np.ndarray):
    x_out = np.zeros((x_in.shape[0] * 2, x_in.shape[1] * 2))
    x_in_pad = np.zeros((x_in.shape[0] + 1, x_in.shape[1] + 1))
    x_in_pad[0 : x_in.shape[0], 0 : x_in.shape[1]] = x_in
    h, w = x_in.shape
    i, j = np.meshgrid(np.arange(h * 2), np.arange(w * 2), indexing="ij")

    x = i / 2
    y = j / 2

    x0 = np.floor(x).astype(int)
    y0 = np.floor(y).astype(int)
    x1 = np.clip(x0 + 1, 0, h - 1)
    y1 = np.clip(y0 + 1, 0, w - 1)
    x0 = np.clip(x0, 0, h - 1)
    y0 = np.clip(y0, 0, w - 1)

    dy = y - np.floor(y)
    dx = x - np.floor(x)

    x_out = (
        x_in[x0, y0] * (1 - dx) * (1 - dy)
        + x_in[x1, y0] * dx * (1 - dy)
        + x_in[x0, y1] * (1 - dx) * dy
        + x_in[x1, y1] * dx * dy
    )

    return x_out
```

```python
print("Original:")
print(orig_2_2)
print("Bilinear:")
print(bilinear(orig_2_2))
```
